$x'=k(x)*x$

Мягкая модель Мальтуса с зависящим от населения коэффициентом размножения

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def dydt(y, t):
    a=0.6
    b=0.1
    return a*y-b*y*y
```

```python
t = np.linspace (0, 10, 100)
```

```python
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten() 
```

```python
%matplotlib inline 

plt.plot(t, solve, label = r"$y'=ay-by^2$")
plt.grid(True)
plt.legend()
plt.show()
```

$y'=y-y^2-c$

Жесткая модель рыболовства с интенсивностью с, где y-количество рыб в озере или мировом океане

```python
def dydt(y, t):
    c=0.05
    return y-y*y-c
```

```python
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten() 
```

```python
t = np.linspace (0, 10, 100)
```

```python
%matplotlib inline 

plt.plot(t, solve, label = r"$y'=y-y^2-c$")
plt.grid(True)
plt.legend()
plt.show()
```

$x'=x-g(x)*x-c$

Мягкая модель рыболовства,где критические значения с являются максимумом функции g(x)x

```python
def dydt(y, t):
    a=0.6
    b=0.3
    c=0.2
    return y-(a-b*y)*y-c
```

```python
t = np.linspace (0, 10, 100)
```

```python
%matplotlib inline 

plt.plot(t, solve, label = r"$y'=y-(a-by)y-c$")
plt.grid(True)
plt.legend()
plt.show()
```

$x'=-kx$

Жесткая модель населения Земли (модель Мальтуса)

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
def dydt(y, t):
    k=0.42
    return -k * y
```

```python
t = np.linspace (0, 10, 100)
```

```python
solve = odeint(dydt, y0=1, t=t)
solve = solve.flatten() 
```

```python
%matplotlib inline 

plt.plot(t, solve, label = r"$y'=-ky$")
plt.grid(True)
plt.legend()
plt.show()
```
