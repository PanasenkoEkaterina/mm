```python
from scipy.integrate import odeint
def dydt(y,t):
    k=0.01
    return k*y
t=np.linspace(1615, 2021, 30)
print(t)
```

```python
solve=odeint(dydt,y0=7000,t=years)
solve=solve.flatten()
%matplotlib inline

plt.plot(years, people, label=r"Население Воронежа")
plt.plot(years, solve, label=r"Жесткая модель Мальтуса")
plt.grid(True)
plt.legend()
plt.show()
```

```python
def p(y):
    a=50000000
    b=350
    return a - b*y
def dydt2(y,t):
    return p(y)*y
```

```python
solve2=odeint(dydt2,y0=7000,t=years)
solve2=solve2.flatten()
%matplotlib inline

plt.plot(years, people, label=r"Население Воронежа")
plt.plot(years, solve2, label=r"Мягкая модель Мальтуса")
plt.grid(True)
plt.legend()
plt.show()
```
