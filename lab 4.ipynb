```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import Ridge

x1=np.arange(1,11)
y1=[1,-43,11,20,55,83,-5,-8,-1,27]

plt.plot(x1,y1,label='katrisha')
plt.grid(True)
plt.legend
plt.show
```

```python
X=np.vstack([x1,np.ones(len(x1))])
X=X.T
print(X)
```

```python
plt.axis([1,10,-30,100])
plt.title('interpol')
plt.plot (x1,y1,'ro')
lsq=np.linalg.lstsq(X,y1,rcond=None)
a,b=lsq[0]
plt.plot(x1, a*x1+b)
plt.show()
```

```python
def f(x,a,b):
    return a*x**2+b*x
(a,b), _ = curve_fit(f,x1,y1)

%matplotlib inline 
plt.plot(x1,y1,".")
plt.plot(x1,f(x1,a,b))
```

```python
x_m=x1[:,np.newaxis]
model=make_pipeline(PolynomialFeatures(1),Ridge())
model.fit(x_m,y1)
x_long=np.linspace(-43,83)[:,np.newaxis]
y_p=model.predict(x_long)
plt.plot(x_long,y_p)
```
